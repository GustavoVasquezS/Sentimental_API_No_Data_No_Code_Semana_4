```python
import pandas as pd

# 1. Cargar archivos con sus codificaciones correctas
df_b2w = pd.read_csv('B2W-Reviews01.csv', sep=None, engine='python', encoding='utf-8-sig')
df_ml = pd.read_csv('reviews_final.csv', sep=None, engine='python', encoding='latin-1')

# 2. Estandarizar B2W
# Mapeamos 'site_category_lv1' como la categoría del producto
df_b2w_std = pd.DataFrame({
    'stars': df_b2w['overall_rating'],
    'review_title': df_b2w['review_title'],
    'review_body': df_b2w['review_text'],
    'language': 'pt',
    'product_category': df_b2w['site_category_lv1']
})

# 3. Estandarizar Mercado Livre (Corregido)
# Ahora usamos 'product_url' para llenar 'product_category'
df_ml_std = pd.DataFrame({
    'stars': df_ml['rating'],
    'review_title': "",
    'review_body': df_ml['content'],
    'language': 'pt',
    'product_category': df_ml['product_url'] # <--- Corregido aquí
})

# 4. Concatenar ambos DataFrames
df_final = pd.concat([df_b2w_std, df_ml_std], ignore_index=True)

# 5. Limpieza final
df_final['stars'] = pd.to_numeric(df_final['stars'], errors='coerce').fillna(0).astype(int)
df_final = df_final.dropna(subset=['review_body'])

# 6. Guardar el archivo consolidado
df_final.to_csv('reviews_consolidado_perfecto.csv', index=False, encoding='utf-8-sig', sep=';')

print("✅ ¡Archivo corregido y guardado con éxito!")
print(f"Total de registros: {len(df_final)}")
display(df_final.tail()) # Mostramos el final para ver los datos de Mercado Livre
```
